```python
from collections import Counter
all_skills = []
for skills in df["Skills_Required"]:
    skills_list = skills.split(",")
    
    for skill in skills_list:
        all_skills.append(skill.strip())
        
skill_counts = Counter(all_skills)
skill_counts.most_common(20)
```

```python
df["Job_Title"].value_counts().head(20)
```

```python
data_roles =  df[df["Job_Title"].isin(["Data Analyst","Data Scientist","Data Engineer","Business Analyst"])]
data_roles.shape
```

```python
data_roles["Job_Title"].value_counts()
```

```python
from collections import Counter
data_skills = []
for skills in data_roles["Skills_Required"]:
    skills_list = skills.split(",")
    
    for skill in skills_list:
        data_skills.append(skill.strip())

data_skill_counts = Counter(data_skills)
data_skill_counts.most_common(20)
```

```python
skill_salary = []
for skills, salary in zip(data_roles["Skills_Required"],data_roles["Salary_LPA"]):
    for skill in skills.split(","):
        skill_salary.append([skill.strip(),salary])
skill_salary[:10 ]
    

```

```python
skill_salary_df = pd.DataFrame(skill_salary, columns = ["skill","salary"])
skill_salary_df.head(10)
```

```python
skill_salary_df.groupby("skill")["salary"].mean().sort_values(ascending=False)
```

```python
skill_salary_df.groupby("skill")["salary"].count()
```

```python
top_skills_salary = (skill_salary_df.groupby("skill")["salary"].mean().sort_values(ascending=False).head(10))
top_skills_salary
```

```python
import matplotlib.pyplot as plt
top_demanded =  data_skill_counts.most_common(10)

skills = [item[0] for item in top_demanded]
counts = [item[1] for item in top_demanded]

plt.figure(figsize=(10,5))
plt.bar(skills, counts,color = "purple")
plt.xticks(rotation = 90)
plt.title("Top 10 Demanded Skills in Data Roles")
plt.show()

```

```python
skill_demand = pd.DataFrame(data_skill_counts.most_common(20),columns = ["skill","demand_count"])
skill_salary = (skill_salary_df.groupby("skill")["salary"].mean().reset_index())
skill_analysis = pd.merge(skill_demand,skill_salary,on="skill")
skill_analysis
```

```python
skill_analysis["opportunity_score"] = skill_analysis["demand_count"] * skill_analysis["salary"]
skill_analysis.sort_values(by="opportunity_score",ascending=False).head(10)
```
